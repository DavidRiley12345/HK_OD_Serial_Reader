```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
with open('/home/ppe/d/driley/OD_testing/NOISE_SCAN/NOISE_SCAN_270924.txt') as f:
    data = f.readlines()
print(data[0])
data = eval(data[0])
```

```python
data_list = {}
for ch_no, values in data.items():
    print(ch_no, values)
    DAC_values = []
    rates = []
    for DAC_value, count in values.items():
        DAC_values.append(DAC_value)
        # convert counts to hertz
        readouts = int(count)/4
        rate = readouts/8
        rates.append(rate)
    data_list[ch_no] = [DAC_values, rates]



for ch_no, values in data_list.items():
    plt.plot(values[0], values[1], label='Channel {}'.format(ch_no))
plt.xlabel('DAC Value')
plt.ylabel('Rate / hz')
plt.legend()
```
